To create a Dataframe

```python
import pandas as pd

data={
    "calories":[420,380,390],
    "duration":[50,40,45]

}

#load data into a dataframe object:
df = pd.DataFrame(data)

print(df)
```

To locate the rows in Dataframe

```python
print(df.loc[[0,1]])
```

To name the index

```python
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}

df=pd.DataFrame(data, index=["day1","day2","day3"])

print(df)
```

To locate Named index


```python
print(df.loc["day2"])
```
